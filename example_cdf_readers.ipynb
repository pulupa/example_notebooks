# Examples of how to perform basic operations (read, print, extract) on a CDF format file using three different Python libraries

This notebook presents code examples of how to perform basic operations (read, print, extract) on a [CDF format file](https://cdf.gsfc.nasa.gov/) format file.
The following Python libraries are tested:
- [spacepy](https://pypi.org/project/spacepy)
- [cdflib](https://pypi.org/project/cdflib)
- [pycdfpp](https://pypi.org/project/pycdfpp)

IMPORTANT: 
- Make sure these libraries are installed locally before running this notebook
- Make sure the CDF file to read has been already downloaded (here we use by default the file solo_l3_rpw-tnr-surv-flux_20240324_v02.cdf, which can be retrieved for example from https://cdaweb.gsfc.nasa.gov/pub/data/solar-orbiter/rpw/science/l3/tnr-surv-flux/2024/solo_l3_rpw-tnr-surv-flux_20240324_v02.cdf)

See at the bottom of the notebook for an additional list of tools to handle CDF format files.

```python
# First, you need to set the CDF file to be read in the notebook
cdf_file_to_read = 'solo_L3_rpw-tnr-surv-flux_20240323_V02.cdf'
import os
# Check if the file does exist
if os.path.isfile(cdf_file_to_read):
    print(f"{cdf_file_to_read} has been found")
else:
    raise FileNotFoundError(f"{cdf_file_to_read} not found!")
```

# 1. spacepy

Website: https://pypi.org/project/spacepy
Version tested: 0.7.0

```python
try:
    from spacepy.pycdf import CDF
except ImportError:
    print("Are you sure spacepy has been installed correctly?")
    raise 
else:
    print("spacepy imported")
```

```python
# 1.1 Open the CDF file
try:
    cdf = CDF(cdf_file_to_read)
except spacepy.pycdf.CDFError:
    print(f'{cdf_file_to_read} cannot be read!')
    raise
else:
    print(f'{cdf_file_to_read} opened')
```

```python
# 1.2 Returns basic CDF information
print('BASIC CDF INFOS ##############################################################')
print()
print(cdf)
```

```python
# 1.3 Retrieve and print all the global attributes of the file (header metadata) 
print('GLOBAL ATTRIBUTES ##############################################################')
print()
gattrs = cdf.attrs 
print(gattrs) # print on the screen all the global attributes
```

```python
# 1.4 Retrieve and print the list of variables in the file
print('LIST OF VARIABLES ##############################################################')
print()
cdf_vars = cdf.keys()
print(cdf_vars)
```

```python
# 1.5 Retrieve and print the attributes of a given variable (here PSD_FLUX)
print('PSD_FLUX VARIABLE ATTRIBUTES ##############################################################')
print()
vattrs = cdf['PSD_FLUX'].attrs
print(vattrs) # print the attributes of the variable PSD_FLUX
```

```python
# 1.6 Copy values of a given variable (here PSD_FLUX) in a new variable named flux
# and print the type and shape of flux 
print('PSD_FLUX VARIABLE TYPE AND SHAPE ##############################################################')
print()
flux=cdf['PSD_FLUX'][...] 
print('PSD_FLUX: ', type(flux), flux.shape)
```

```python
# 1.7 Close the file (important!)
cdf.close()
```

# 2. cdflib

Website: https://pypi.org/project/cdflib
Version tested: 1.3.10

```python
try:
    import cdflib
except ImportError:
    print("Are you sure cdflib has been installed correctly?")
    raise 
else:
    print("cdflib imported")
```

```python
# 1.1 Open the CDF file
try:
    cdf = cdflib.CDF(cdf_file_to_read)
except (FileNotFoundError, OSError):
    print(f'{cdf_file_to_read} cannot be read!')
    raise
else:
    print(f'{cdf_file_to_read} opened')
```

```python
# 1.2 Returns basic CDF information
print('BASIC CDF INFOS ##############################################################')
print()
print(cdf.cdf_info()) 
```

```python
# 1.3 Retrieve and print all the global attributes of the file (header metadata) 
print('GLOBAL ATTRIBUTES ##############################################################')
print()
gattrs = cdf.globalattsget()
print(gattrs)
```

```python
# 1.4 Retrieve and print the list of variables in the file
print('LIST OF VARIABLES ##############################################################')
print()
variables = cdf.cdf_info().zVariables
print(variables)
```

```python
# 1.5 Retrieve and print the attributes of a given variable (here PSD_FLUX)
print('PSD_FLUX VARIABLE ATTRIBUTES ##############################################################')
print()
vattrs = cdf.varattsget('PSD_FLUX')
print(vattrs)
```

```python
# 1.6 Copy values of a given variable (here PSD_FLUX) in a new variable named flux
# and print the type and shape of flux
print('PSD_FLUX VARIABLE TYPE AND SHAPE ##############################################################')
print()
flux=cdf.varget('PSD_FLUX')
print('PSD_FLUX: ', type(flux), flux.shape)
```

```python
# 1.7 Close the file 
# cdflib does not have cdf.close() method
```

# 3. pycdfpp 

Website: https://pypi.org/project/pycdfpp/
Version tested: 0.9.3

```python
try:
    import pycdfpp
except ImportError:
    print("Are you sure pycdfpp has been installed correctly?")
    raise 
else:
    print("pycdfpp imported")
```

```python
# 1.1 Open the CDF file
try:
    cdf=pycdfpp.load(cdf_file_to_read)
except:
    print("Are you sure pycdfpp has been installed correctly?")
    raise 
else:
    print(f'{cdf_file_to_read} opened')
```

```python
# 1.2 Returns basic CDF information
print('BASIC CDF INFOS ##############################################################')
print()
```

```python
# 1.3 Retrieve and print all the global attributes of the file (header metadata) 
print('GLOBAL ATTRIBUTES ##############################################################')
print()
gattrs = cdf.attributes
print(gattrs)
```

```python
# 1.4 Retrieve and print the list of variables in the file
print('LIST OF VARIABLES ##############################################################')
print()
variables = list(cdf)
print(variables)
```

```python
# 1.5 Retrieve and print the attributes of a given variable (here PSD_FLUX)
print('PSD_FLUX VARIABLE ATTRIBUTES ##############################################################')
print()
vattrs = cdf['PSD_FLUX'].attributes
print(vattrs)
```

```python
# 1.6 Copy values of a given variable (here PSD_FLUX) in a new variable named flux
# and print the type and shape of flux
print('PSD_FLUX VARIABLE TYPE AND SHAPE ##############################################################')
print()
flux = cdf['PSD_FLUX'].values
print('PSD_FLUX: ', type(flux), flux.shape)
```

```python
# 1.7 Close the file
# pycdfpp does not have cdf.close() method
```
